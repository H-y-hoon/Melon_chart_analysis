```python
title = []
artist = []
id = []
lyric = []

for m in range(1, 13):
    # time.sleep(5)
    hdr={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36",
    "Cookie": "__T_=1; __T_=1; PCID=16685154266303576823217; PC_PCID=16685154266303576823217; __T_=1; POC=WP10; mainPop=2022%3A12%3A05%2023%3A59%3A59; wcs_bt=s_f9c4bde066b:1670219226; _T_ANO=ckrX6AALkw5+A2/S6yICwSI1VidQcDBRabvcYWS1pVSNmnKOQGxc5dA6d1qQRWxgslvI+48D7XHMfE9ap0uXPZRX6i/JXKJtt03ZLhEsNGhKS55xBUnIuzDO8DQtRXvqF1Rhm+ZAusIiz/cYtUTpNT9W/MUiOea70aJQQLq+zRNIwdkkzno4jyF0soUfmG0r9G2c7AkjCp0WyzSeCbQ6/VqMDPgrie4szV571rvVEECkVTTjyXfX1qgoXXdg7UGE+eXfbENVyfJRDvXxhRbarmxuY3NhkY0ATUrasdcAXsV7FdkqkVx2FTcI5FVm3PVnAdTFrmXek0UlcVkGTQwMvg=="}
    if m < 10:
        data = requests.get("https://www.melon.com/chart/search/list.htm?chartType=MO&age=2010&year=2019&mon=0" + str(m)+ "&classCd=DM0000&moved=Y", headers=hdr)
    else:
        data = requests.get("https://www.melon.com/chart/search/list.htm?chartType=MO&age=2010&year=2019&mon=" + str(m) + "&classCd=DM0000&moved=Y", headers=hdr)
    soup = BeautifulSoup(data.content, 'html.parser')
    # 제목 추출
    text_data = list(soup.find_all('strong'))
    # len(text_data)
    for t in text_data[0:-1]:
        title.append(t.text)
    print(len(title))
    # 가수 추출
    text_data = list(soup.find_all(class_="checkEllipsis"))
    # len(text_data)
    for i in text_data:
        artist.append(i.text)
    print(len(artist))
    # songID 추출
    text_data = list(soup.find_all("button"))
    for m in text_data:
        if m.text == '재생':
            print(m['data-song-no'])
            hdr = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
            lyric_data = requests.get("https://www.melon.com/song/detail.htm?songId=" + str(m['data-song-no']), headers=hdr)
            lyric_soup = BeautifulSoup(lyric_data.content, 'html.parser')

            text_data = list(lyric_soup.find_all(id="d_video_summary"))
            if len(text_data) == 0:
                lyric.append('None')
            else:
                l = text_data[0].text.replace('\r', '').replace('\t', '')
                lyric.append(l)
            id.append(m['data-song-no'])

import pandas as pd

df = pd.DataFrame({"songID": id, "제목": title, "가수": artist, "가사": lyric})
df.to_csv("melon_2019.csv", encoding='utf-8')
```
