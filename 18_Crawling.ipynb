```python
title = []
artist = []
id = []
lyric = []

for m in range(1, 13):
    # time.sleep(5)
    hdr={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36",
    "Cookie": "__T_=1; __T_=1; PCID=16694648515334418466078; PC_PCID=16694648515334418466078; __T_=1; POC=WP10; wcs_bt=s_f9c4bde066b:1669467732; _T_ANO=Emma7KKrBkrgu9MTPGoAE0sSUhAGqScTNH+rW6p4A0j3sFwHyq+YrYr6IcqMEHLyIQw0olgO0inzjHvORhq2/d/iAf+3TLOkJDKrYAa+MopgHsY6V2H3QXit1r0Art3wy8/hLTBYCn2MgwY6in4OiJ0ecONJ49YE40+OYvFxZ6KjwjieoHg0BiVUpBe0rCLNz4ZHPuTLN+EOBrDV7FkfeTYz1x1susGhfCpb85AU/SZxANBIL3xW9iyGHxNs6melTo7AaqawAA1gS5RAV8WClWiLFWk39fotOWKzB/wefNVE8e0yJOl95jjJR9zRGvKmgWchySrDLbttuQsgXPS0Fw=="}
    if m < 10:
        data = requests.get("https://www.melon.com/chart/search/list.htm?chartType=MO&age=2010&year=2018&mon=0" + str(m)+ "&classCd=DM0000&moved=Y", headers=hdr)
    else:
        data = requests.get("https://www.melon.com/chart/search/list.htm?chartType=MO&age=2010&year=2018&mon=" + str(m) + "&classCd=DM0000&moved=Y", headers=hdr)
    soup = BeautifulSoup(data.content, 'html.parser')
    # 제목 추출
    text_data = list(soup.find_all('strong'))
    # len(text_data)
    for t in text_data[0:-1]:
        title.append(t.text)
    print(len(title))
    # 가수 추출
    text_data = list(soup.find_all(class_="checkEllipsis"))
    # len(text_data)
    for i in text_data:
        artist.append(i.text)
    print(len(artist))
    # songID 추출
    text_data = list(soup.find_all("button"))
    for m in text_data:
        if m.text == '재생':
            print(m['data-song-no'])
            hdr = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
            lyric_data = requests.get("https://www.melon.com/song/detail.htm?songId=" + str(m['data-song-no']), headers=hdr)
            lyric_soup = BeautifulSoup(lyric_data.content, 'html.parser')

            text_data = list(lyric_soup.find_all(id="d_video_summary"))
            if len(text_data) == 0:
                lyric.append('None')
            else:
                l = text_data[0].text.replace('\r', '').replace('\t', '')
                lyric.append(l)
            id.append(m['data-song-no'])

import pandas as pd

df = pd.DataFrame({"songID": id, "제목": title, "가수": artist, "가사": lyric})
df.to_csv("melon_2018.csv", encoding='utf-8')
```
