```python
####텍스트 읽어오기####
f = open("/content/drive/MyDrive/Output18_1.txt",'rt',encoding='utf-8')
lines = f.readlines()
line = []
for i in range(len(lines)):
    line.append(lines[i])
f.close()
```

```python
print(line)
```

```python
###특수문자 제거하기###
##한국어##
import re
compile = re.compile("[^ ㄱ-ㅣ가-힣]+")
for i in range(len(line)):

    a = compile.sub("",line[i])
    line[i] = a
print(line)
```

```python
Okt = Okt()
line_ex = line[0]
print("형태소 단위로 문장 분리")
print("----------------------")
print(Okt.morphs(line[0]))
print(" ")
print("문장에서 명사 추출")
print("----------------------")
print(Okt.nouns(line[0]))
print(" ")
print("품사 태킹(PoS)")
print("----------------------")
print(Okt.pos(line[0]))
```

```python
# 노래별 명사추출
dfdf = []
for idx, text in enumerate(line):
  dfdf.append(Okt.nouns(line[idx]))
```

```python
# 노래별 이라는 구분을 짓는 이중 리스트 제거
dfdf2 = [y for x in dfdf for y in x]
dfdf2
```

```python
# 불용어 정리
stop_words = ['나','너','네', '내', '널', '날', '너를', '나를', '우리', 
             '와', '이', '있', '하', '것', '들', '그', '되', '수', '이',
             '보', '않', '나', '사람', '주', '아니', '등', '같', '나오',
             '가지', '씨', '시키', '만들', '지금', '생각하', '그러', '속',
             '하나', '집', '살', '모르', '적', '월', '데', '자신', '안',
             '어떤', '때', '년', '가', '한', '지', '경우', '명', '생각',
             '대하', '시간', '오', '그녀' ,'말', '다시' ,'일', '이런',
             '그렇', '앞', '위하', '보이', '때문', '번', '그것', '두',
             '다른', '말하', '어떻', '알', '여자', '그러나', '개', '받',
             '전', '못하', '들', '들', '일', '사실', '그런', '이렇', '또',
             '점', '문제', '싶', '더', '말', '많', '그리고', '잘', '좋',
             '잘', '놓', '나는', '너는', '올', '난', '우린', '내가','넌','걸','거','게','이제','못','왜']
```

```python
# 불용어 제거
for stop_word in stop_words:
  while stop_word in dfdf2:
    dfdf2.remove(stop_word)
```
